# Load Images + JSON Labels + Target Masks

```python
import json
import numpy as np
import cv2

def json_to_mask(json_path, img_shape):

    mask = np.zeros(img_shape[:2], dtype=np.uint8)

    with open(json_path) as f:
        data = json.load(f)

    h, w = img_shape[:2]

    features = data['features']['xy']

    for feature in features:

        wkt = feature['wkt']
        damage = feature.get('properties', {}).get('subtype', 'no-damage')

        # Class mapping
        class_map = {
            'no-damage': 1,
            'minor-damage': 2,
            'major-damage': 3,
            'destroyed': 4
        }

        value = class_map.get(damage, 0)

        if "POLYGON" in wkt:

            pts = wkt.replace('POLYGON ((', '').replace('))', '')
            pts = pts.split(',')

            polygon = []

            for pt in pts:
                x, y = map(float, pt.strip().split())
                polygon.append([int(x), int(y)])

            polygon = np.array([polygon], dtype=np.int32)

            cv2.fillPoly(mask, polygon, value)

    return mask
```

```python
import os
import cv2
import numpy as np

def load_dataset(img_dir, json_dir, img_size):

    X = []
    y = []

    img_files = os.listdir(img_dir)

    for img_file in img_files:

        if not img_file.endswith('.png'):
            continue

        img_path = os.path.join(img_dir, img_file)
        json_path = os.path.join(json_dir, img_file.replace('.png', '.json'))

        img = cv2.imread(img_path)

        if img is None:
            continue

        img = cv2.resize(img, (img_size, img_size))
        img = img / 255.0

        if not os.path.exists(json_path):
            continue

        mask = json_to_mask(json_path, img.shape)

        mask = cv2.resize(
            mask,
            (img_size, img_size),
            interpolation=cv2.INTER_NEAREST
        )

        X.append(img)
        y.append(mask)

    X = np.array(X)
    y = np.array(y)

    return X, y
```

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D
from tensorflow.keras.layers import UpSampling2D, concatenate, BatchNormalization
from tensorflow.keras.models import Model


def conv_block(x, filters):

    x = Conv2D(filters, 3, activation='relu', padding='same')(x)
    x = BatchNormalization()(x)

    x = Conv2D(filters, 3, activation='relu', padding='same')(x)
    x = BatchNormalization()(x)

    return x


inputs = Input((256,256,3))

# Encoder
c1 = conv_block(inputs, 64)
p1 = MaxPooling2D()(c1)

c2 = conv_block(p1, 128)
p2 = MaxPooling2D()(c2)

c3 = conv_block(p2, 256)
p3 = MaxPooling2D()(c3)

# Bridge
c4 = conv_block(p3, 512)

# Decoder
u1 = UpSampling2D()(c4)
u1 = concatenate([u1, c3])
c5 = conv_block(u1, 256)

u2 = UpSampling2D()(c5)
u2 = concatenate([u2, c2])
c6 = conv_block(u2, 128)

u3 = UpSampling2D()(c6)
u3 = concatenate([u3, c1])
c7 = conv_block(u3, 64)

# Final output (5 classes)
outputs = Conv2D(5, (1,1), activation='softmax')(c7)

model = Model(inputs, outputs)
```

```python
# Full compile block 
```

```python
from tensorflow.keras.optimizers import Adam

model.compile(
    optimizer=Adam(learning_rate=0.0001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# AI Disaster Damage Assessment Detection System
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import load_model

# Load trained model
model = load_model("damage_model.h5")


# -----------------------------
# Class names
# -----------------------------
class_names = {
    0: "Background",
    1: "No Damage",
    2: "Minor Damage",
    3: "Major Damage",
    4: "Destroyed"
}


# -----------------------------
# Heatmap Generator
# -----------------------------
def create_damage_map(pred):

    pred_mask = np.argmax(pred, axis=-1)[0]

    heatmap = np.uint8(pred_mask * 60)
    heatmap = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)

    return pred_mask, heatmap


# -----------------------------
# Detection Function
# -----------------------------
def detect_and_display(before_path, after_path, model):

    # Load images
    before = cv2.imread(before_path)
    after = cv2.imread(after_path)

    before = cv2.cvtColor(before, cv2.COLOR_BGR2RGB)
    after = cv2.cvtColor(after, cv2.COLOR_BGR2RGB)

    # Resize for model
    after_input = cv2.resize(after, (256,256))
    after_input = after_input / 255.0
    after_input = np.expand_dims(after_input, axis=0)

    # Prediction
    pred = model.predict(after_input)

    # Heatmap
    pred_mask, heatmap = create_damage_map(pred)

    # Resize heatmap
    heatmap = cv2.resize(heatmap, (after.shape[1], after.shape[0]))

    # Overlay
    overlay = cv2.addWeighted(after, 0.7, heatmap, 0.3, 0)

    # Damage statistics
    unique, counts = np.unique(pred_mask, return_counts=True)

    total_pixels = pred_mask.size

    print("\nDamage Report:")
    for u, c in zip(unique, counts):
        percent = (c / total_pixels) * 100
        print(f"{class_names[u]}: {percent:.2f}%")

    damage_pixels = np.count_nonzero(pred_mask)
    damage_percent = (damage_pixels / total_pixels) * 100

    # Plot
    plt.figure(figsize=(18,5))

    plt.subplot(1,4,1)
    plt.imshow(before)
    plt.title("Before Disaster")

    plt.subplot(1,4,2)
    plt.imshow(after)
    plt.title("After Disaster")

    plt.subplot(1,4,3)
    plt.imshow(heatmap)
    plt.title("Damage Heatmap")

    plt.subplot(1,4,4)
    plt.imshow(overlay)
    plt.title(f"Overlay Damage {damage_percent:.2f}%")

    plt.show()

    print(f"\nTotal Damage Area: {damage_percent:.2f}%")
```
